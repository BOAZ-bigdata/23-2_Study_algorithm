```python
# 11724번 연결요소의 개수

import sys

a ,b = map(int,input().split())
arr = [[] for i in range(a+1)]
for i in range (b):
    n, m = map(int, sys.stdin.readline().rstrip().split())
    arr[n].append(m)
    arr[m].append(n)
vst = []
cnt = 0

for i in range(1, a+1):
    if i in vst:
        continue
    else:
        vst.append(i)
        for s in vst:
            for j in arr[s]:
                if j not in vst:
                    vst.append(j)
        cnt += 1
    
print(cnt)
```

```python
## 2023번 신기한 소수

import math

arr = [[2, 3, 5, 7]]
for i in range (int(input())-1):
    arr.append([])
    for n in arr[i]:
        for s in range(1,10,2):
            t = (10*n)+s
            
            sn = True
            for s in range(3, math.ceil(t**(1/2))+1, 2):
                if t/s == t//s:
                    sn = False
                    break
                    
            if sn == True:
                arr[-1].append(t)
                
for i in arr[-1]:
    print(i)
    
## 굳이 체까지 안 써도 풀립니다, 수가 그렇게 안 커서요
```

```python
## 2178번 미로탐색

import sys

h, w = map(int,sys.stdin.readline().rstrip().split())
arr = []
k = 2
for _ in range (h):
    arr.append(list(map(int,sys.stdin.readline().rstrip())))

arr[0][0] = k
while arr[h-1][w-1] == 1:
    for hi in range (h):
        for wi in range (w):
            if arr[hi][wi] == k:
                if hi < h-1:
                    if arr[hi+1][wi] != 0:
                        arr[hi+1][wi] = k+1
                if wi < w-1:
                    if arr[hi][wi+1] != 0:
                        arr[hi][wi+1] = k+1
                if hi > 0:
                    if arr[hi-1][wi] != 0:
                        arr[hi-1][wi] = k+1
                if wi > 0:
                    if arr[hi][wi-1] != 0:
                        arr[hi][wi-1] = k+1
    k+=1
    
print (k-1)

## 어디선가 봤던 것 같은데.. 노가다로 짜고 나서야 [0 1 0 -1] [1 0 -1 0] 을 써도 될 것 같다는 생각이 들었습니다.
```

```python
## 2805번 나무자르기

import sys

##sys.stdin.readline().rstrip()

n, m = map(int,sys.stdin.readline().rstrip().split())
trr = list(map(int,sys.stdin.readline().rstrip().split()))

start = -1
end = max(trr)
go = False

while (end - start) > 1 or go == False:
    mid = (start+end)//2
    get = 0
    for i in range (n):
        if mid >= trr[i]:
            continue
        elif mid < trr[i]:
            get += trr[i]-mid
    
    if get>m:
        start = mid
        go = True
    elif get<m:
        end = mid
        go = False
    else:
        break

print (mid)

## 교재 문제 너무 어려워서... 다른 이진탐색 풀어왔습니다 ㅠ
```
